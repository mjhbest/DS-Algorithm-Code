```python
import requests
from bs4 import BeautifulSoup
# Python3 샘플 코드 #

url = 'http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getUltraSrtFcst'
params ={'serviceKey' : 'MIe3DGXhgZez8IKsUCzyozfnp9dI8s5CTUpKRvta07JqEs7uLA3tA58Io9TqmV0z8uF1g9rgeWsTDiuSqJs3MA==',
         'pageNo' : '1', 'numOfRows' : '1000', 'dataType' : 'XML', 'base_date' : '20230612', 'base_time' : '0600', 'nx' : '37', 'ny' : '127' }

response = requests.get(url, params=params)
print(response.content)

bsObject = BeautifulSoup(response.content, "lxml-xml")
```

```python
import pandas as pd

categories = []
dates = []
times = []
values = []

items = bsObject.find_all('item')

for item in items:
    categories.append(item.category.text)
    dates.append(item.fcstDate.text)
    times.append(item.fcstTime.text)
    values.append(item.fcstValue.text)

df = pd.DataFrame({
    'category' : categories,
    'date' : dates,
    'time' : times,
    'value' : values
})

df.to_csv('기상청_날씨단기예보.csv')
```
